

1.DATA CLEANING








```python
# checking missing values
df[df.isnull().any(axis = 1)]
```

The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region

1 : "Bejaia Region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
# dividing our into two region and creating a new column
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
# checking missing valuess
df.isnull().sum()
```

```python
# removing null values
# we have dropped those rows which are having null values
# and reset the index so two rows are removed
df = df.dropna().reset_index(drop = False)
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# we will remove this row because it have no data
df = df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# fix spaces in column names
# strip() fix leading or trailing spaces
df.columns = df.columns.str.strip()
df.columns
```

```python
# changing important column as integer type
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
df.info()
```

```python
# leaving classes we will covert all other remaining columns into float dtype
objects = [features for features in df.columns if df[features].dtype == 'O' ]
```

```python
objects
```

```python
for i in objects:
  if i!='Classes':
    df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['index'],axis =1)
```

```python
# converting to this new dataset which will be cleaned
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index = False)
```


2.EXPLORATORY DATA ANALYSIS



```python
df_copy = df.drop(['day','month','year','index'],axis = 1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
# encoding of the categories in classes
import numpy as np
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python

## plot desnsity for all features
import matplotlib.pyplot as plt
plt.style.use('seaborn')
df_copy.hist(bins = 50,figsize = (20,15))
plt.show()
```

```python
percentage = df_copy['Classes'].value_counts(normalize = True)*100
percentage
```

```python
classlabels = ['fire','not fire']
plt.figure(figsize = (10,7))
plt.pie(percentage,labels = classlabels,autopct = '%1.1f%%')
plt.title("pie chart of classes")
plt.show()
```
